```python
import pandas as pd

# URL of the table
url = "https://pokemondb.net/pokedex/all"

# Read all HTML tables on the page
tables = pd.read_html(url)

# The first table is the main Pok√©mon table
df = tables[0]

# Save as CSV
df.to_csv("pokemon_all_gen.csv", index=False)
print("CSV saved as 'pokemon_all.csv'")

# Optional: see the first few rows
df.head()
```
