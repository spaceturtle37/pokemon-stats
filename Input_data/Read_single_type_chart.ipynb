# "https://pokemondb.net/type"

```python
import pandas as pd

url = "https://pokemondb.net/type"

# This will read all HTML tables on the page
tables = pd.read_html(url)
df = tables[0]    # or tables[n] whichever is correct
print(df.shape)
# Optional: see the first few rows
df.head()
```

```python
# fix the shape and row/column labels
# name = df.index.name
name = df.columns.to_list()[0]
print(name)
types = df[name].to_list()
print(types)
df.set_index(name, inplace=True)
df.columns = types
print(df.shape)
df.head()
```

```python
# replace Nans with value of 1
df = df.fillna(1)
df.head()
```

```python
# replace 1/2 with 0.5 
df = df.replace(to_replace='½', value=0.5)
df.head()
```

```python
# ensure all columns are numeric and no strings
df = df.apply(pd.to_numeric)
print(df.dtypes)
df.head()
```

```python
print(df.index.name)
df.index.name = 'attack\\defense'
print(df.index.name)
```

```python
df.to_csv("pokemon_type_chart.csv")
print("Saved as pokemon_type_chart.csv")
```
