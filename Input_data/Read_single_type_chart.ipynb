# "https://pokemondb.net/type"

```python
import pandas as pd

import pokemon_stats.mat_utils as mat
```

```python
url = "https://pokemondb.net/type"

# This will read all HTML tables on the page
tables = pd.read_html(url)
df = tables[0]    # or tables[n] whichever is correct
# print(df.shape)
# Optional: see the first few rows
df.head()
```

# set types as index

```python
df.set_index(df.columns[0], inplace=True)
df.head()
```

# rename columns to match index

```python
df.columns = df.index
df
```

# replace Nans with value of 1

```python
df = df.fillna(1)
df.head()
```

# fix decimal strings into numeric

```python
# replace 1/2 with 0.5 
df = df.replace(to_replace='½', value=0.5)
df.head()
```

# convert columns to numeric type

```python
# ensure all columns are numeric and no strings
df = df.apply(pd.to_numeric)
# print(df.dtypes)
df.head()
```

# rename the top left corner

```python
# print(df.index.name)
df.index.name = 'attack\\defense'
# print(df.index.name)
df
```

# Export

```python
mat.exportdf(df, 'single_single_type_chart')
```
