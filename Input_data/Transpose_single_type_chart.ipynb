# despite the convention adopted by the single type table "https://pokemondb.net/type" 

# they tranpose row and columns for the dual type https://pokemondb.net/type/dual

# so i decided for consistency to transpose the single type table too, so rows will always be defending type

```python
import pandas as pd
df = pd.read_csv('pokemon_type_chart.csv', index_col=0)
df.head()
```

```python
# transpose the table
df = df.T
# fix the defense attack label
df.columns.name = ''
# print(df.index.name)
# only the index name gets carried over, not the column name
df.index.name = 'defense\\attack'
# print(df.index.name)
# print(type_chart.columns)
df.head()
```

```python
# ensure all columns are numeric and no strings
df = df.apply(pd.to_numeric)
print(df.dtypes)
df.head()
```

```python
# save the transpoed and filled in version
file_name = "single_type_chart.csv" 
df.to_csv(file_name)
print("Saved as " + file_name)
```
