```python
import pandas as pd
```

```python
# read data
df = pd.read_csv('dual_type_chart' + '.csv', index_col=0)
df.head()
```

# iterate over single types to generate table for double type vs double type

```python
dual_types = df.columns
n_dual_types = len(dual_types)

single_types = df.index
n_single_types = len(single_types)

df_dual_dual = pd.DataFrame(index=dual_types, columns=dual_types)
df_dual_dual.index.name = 'attack\\defense'
# print(df_dual_dual.head(20))

# dual_types = dual_types[:10]

for atk_type in dual_types:
    atk_type1, atk_type2 = atk_type.split()
    df_dual_dual.loc[atk_type] = df.loc[atk_type.split()].max(axis=0)

df_dual_dual
```

# Export

```python
# save the transpoed and filled in version
file_name = "full_dual_dual_type_chart_from_dual.csv" 
df_dual_dual.to_csv(file_name)
print("Saved as " + file_name)
```
