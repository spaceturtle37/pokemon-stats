# Data extracted from
# https://pokemondb.net/type/dual

```python
import pandas as pd

import pokemon_stats.mat_utils as mat
```

```python
url = "https://pokemondb.net/type/dual"

# This will pull *all* HTML tables on the page
tables = pd.read_html(url)
# Identify which table is the dual‑type matrix you want:
df = tables[0]  # (or whichever index matches)
print(df.shape)
# Preview
df.head()
```

# make the type names the index

```python
# fix the index
name = df.columns.to_list()[0]
# print(name)
df.set_index(name, inplace=True)
print(df.shape)
df.head()
```

# Replace NaN with 1

```python
# replace Nans with value of 1
df = df.fillna(1)
print(df.head(25))
```

# fix fractions to be numerical values

```python
# replace 1/2 with 0.5 and 1/5 with 0.25
df = df.replace(to_replace='½', value=0.5)
df = df.replace(to_replace='¼', value=0.25)
df.head()
```

# remove the rows that are just column names

```python
# remove the unncecessary rows with only labels
df = df[~df.index.duplicated(keep=False)]
print(df.shape)
df.head()
```

# convert all columns to numeric type 

```python
# ensure all columns are numeric and no strings
df = df.apply(pd.to_numeric)
# print(df.dtypes)
df.head()
```

# fix the name at the top left corner

```python
print(df.index.name)
df.index.name = 'defense\\attack'
df.head()
```

# Rename column single types to match single type labels

```python
single_type_index = df.index.str.contains('—')
# print(df.index[single_type_index])
single_types = list(df.index[single_type_index].str.replace('—', '', regex=False))
df.columns = ['Pokemon'] + single_types
df.head()
```

# Turn row labels into tuples

```python
# remove the — character for single types
new_index = df.index.str.replace('—', '', regex=False)
# print(df.index)
# turn into tuples
new_index =  [x.split() for x in new_index]
new_index = [x[0] if len(x)==1 else tuple(x) for x in new_index]
df.index = new_index
df
```

# Export

```python
mat.exportdf(df,"dual_single_type_chart")
```
