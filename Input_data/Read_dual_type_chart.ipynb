```python
import pandas as pd

url = "https://pokemondb.net/type/dual"

# This will pull *all* HTML tables on the page
tables = pd.read_html(url)
# Identify which table is the dual‑type matrix you want:
df = tables[0]  # (or whichever index matches)
print(df.shape)
# Preview
df.head()
```

```python
# fix the index
name = df.columns.to_list()[0]
print(name)
df.set_index(name, inplace=True)
print(df.shape)
df.head()
```

```python
# replace Nans with value of 1
df = df.fillna(1)
print(df.head(50))
```

```python
# replace 1/2 with 0.5 and 1/5 with 0.25
df = df.replace(to_replace='½', value=0.5)
df = df.replace(to_replace='¼', value=0.25)
df.head()
```

```python
# remove the unncecessary rows with only labels
df = df[~df.index.duplicated(keep=False)]
print(df.shape)
df.head()
```

```python
# ensure all columns are numeric and no strings
df = df.apply(pd.to_numeric)
print(df.dtypes)
df.head()
```

```python
print(df.index.name)
df.index.name = 'defense\\attack'
print(df.index.name)
```

```python
df.to_csv("dual_type_chart.csv")
print("Saved as dual_type_chart.csv")
```
