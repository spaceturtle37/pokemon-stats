```python
import pandas as pd

import pokemon_stats.mat_utils as mat
import pokemon_stats.viz as viz

%load_ext autoreload
%autoreload 2
```

```python
# read data
df = mat.importdf('single_single_type_chart')
df.head()
```

# calculate Descriptive Statistics

```python
df2 = mat.calculate_net_statistics(df)
df2
```

# Export

```python
mat.exportdf(df2, 'single_single_type_aggregate_statistics')
```

# Damage In

```python
viz.barcharts(-df2['damage_in_avg'], sort='True', color_by='types')
```

# Damage Out

```python
viz.barcharts(df2['damage_out_avg'], sort='True', color_by='types')
```

# Damage Net

```python
viz.barcharts(df2['damage_net_avg'], sort='True', color_by='types')
```

We can see steel has deals the most net damage and grass is the worst overall

# Visualize Damage Received Scatter

```python
viz.scatter_stats(-df, title='Damage In', ylabel='Damage In', alpha=0.5, color_by='types', sort=True)
```

We can see steel has the least damage taken, followed by ghost no surprise given double 0 damage instances

# Visualize Damage Given Scatter

```python
viz.scatter_stats(df.T, title='Damage Out', ylabel='Damage Out', alpha=0.5, color_by='types', sort=True)
```

We can see ground and rock are the best damage dealers

# look at the matrix minus transpose for net damage = out - in

```python
# by doing this we are comparing how much does A damage B vs B damage A. only relative damage
df3 = df - df.T
df3.index.name = 'attack\\defense'
df3
```

# Net Damage

```python
viz.scatter_stats(df3.T, title='Damage Net', ylabel='Damage Net', alpha=0.5, color_by='types', sort=True)
```

# Color by Defending Pokemon and Add Z-score

```python
viz.hover_scatter(df3.T)
```

We see the ranked results has steel be the most consistent, despite not having the strongest relative damages. Grass is the worst, despite normal being a worst or equal matchup against most things
